```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn3

# --- 2-SET VENN DIAGRAM ---
set_a = {"Python", "Data Analysis", "Machine Learning"}
set_b = {"JavaScript", "Web Dev", "Machine Learning"}

plt.figure(figsize=(8, 5))
v = venn2([set_a, set_b], set_labels=("Set A", "Set B"))

# Customize colors
v.get_patch_by_id("10").set_color("#85B7EB")   # Only A
v.get_patch_by_id("01").set_color("#5DCAA5")   # Only B
v.get_patch_by_id("11").set_color("#AFA9EC")   # Intersection

plt.title("2-Set Venn Diagram")
plt.tight_layout()
plt.savefig("venn2.png", dpi=150)
plt.show()


# --- 3-SET VENN DIAGRAM ---
set_a = {"Python", "Data Analysis", "Machine Learning"}
set_b = {"JavaScript", "Web Dev", "Machine Learning"}
set_c = {"Statistics", "Data Analysis", "Machine Learning"}

plt.figure(figsize=(8, 6))
v = venn3([set_a, set_b, set_c], set_labels=("Set A", "Set B", "Set C"))

# Customize colors
v.get_patch_by_id("100").set_color("#85B7EB")  # Only A
v.get_patch_by_id("010").set_color("#5DCAA5")  # Only B
v.get_patch_by_id("001").set_color("#EF9F27")  # Only C
v.get_patch_by_id("110").set_color("#AFA9EC")  # A ∩ B
v.get_patch_by_id("101").set_color("#9FE1CB")  # A ∩ C
v.get_patch_by_id("011").set_color("#FAC775")  # B ∩ C
v.get_patch_by_id("111").set_color("#D3D1C7")  # A ∩ B ∩ C

plt.title("3-Set Venn Diagram")
plt.tight_layout()
plt.savefig("venn3.png", dpi=150)
plt.show()
```
